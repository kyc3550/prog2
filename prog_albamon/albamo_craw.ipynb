```python
# 먼저, 크롬을 켜 알바몬 사이트에 접속
browser = webdriver.Chrome("./chromedriver.exe")
browser.get("https://www.albamon.com/main")
```

```python
main=browser.find_element_by_id("btn_MType")
main.click()

browser.find_element_by_link_text("경기").click()
browser.find_element_by_link_text("수원시 장안구").click()
time.sleep(1)
browser.find_element_by_id("dev_area_dong_code_B190").click()
s_location=browser.find_element_by_class_name("sch")
s_location.click()
time.sleep(1)
```

```python
base_url = urllib.request.urlopen(browser.current_url).read()
soup = BeautifulSoup(base_url,"html.parser")

ths = []
tds = []
data = []

wrap = soup.find('div',{"class":"gListWrap"})
table = wrap.find('table')
t_head = table.find('thead').find_all('th')
t_body = table.find('tbody')

for th in t_head:
    ths.append(th.get_text())

print("{0}\t\t{1}\t\t{2}\t\t{3}\t\t{4}".format(ths[0],ths[1],ths[2],ths[3],ths[4]))


for tr in t_body.find_all('tr'):
    area = tr.find('td',{'class':'area'}).find('div')
    area.span.decompose()
    subject_name = tr.find('td',{'class':'subject'}).find('p',{'class':'cName'}).text
    subject_ctit = tr.find('td',{'class':'subject'}).find('p',{'class':'cTit'}).text
    pay_money = tr.find('td',{'class':'pay'}).find('p',{'class':'money'}).find("img")["alt"]
    pay_won = tr.find('td',{'class':'pay'}).find('p',{'class':'won'}).text
    
    tds = list(tr.find_all('td'))
    for td in tds:
        time = tds[3].text
        
    print(area.text + subject_name + "( {} )".format(subject_ctit) + "\t" +pay_money +" : "+ pay_won + "\n{} : {}".format(ths[3],time))
    print("--------------------------------------------------------------------------------------------------")


```
