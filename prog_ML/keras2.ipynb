```python
import numpy as np
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import SGD
## y = x/2  - 1 관계를 갖는 가상 데이터를 생성
num_points = 1500 # 1500개의 데이터 쌍
vectors_set = []

for i in range(num_points):
    x= np.random.normal(0, 0.13) # 평균 0, 표준편차 0.13를 갖는 정규분포의 난수 발생
    y= x / 2 - 1 + np.random.normal(0.0, 0.1)
    # 평균 0, 표준편차 0.1를 갖는 정규 분포를 갖는 난수의 편차 추가
    vectors_set.append([x, y])
    
# x, y 데이터 분리
x_data = [v[0] for v in vectors_set]
y_data = [v[1] for v in vectors_set]
# 데이터 분포 그림 그리기
plt.plot(x_data, y_data, 'ro', label='Original data')
```

```python
# 학습 구조 모델링
model = Sequential()

# 입력 x의 차원은 1, 출력 y의 차원도 1. 선형 회귀이므로 activation은 'linear'
model.add(Dense(1, input_dim=1, activation='linear'))

# sgd는 경사 하강법을 의미. 학습속도(learning rate, lr)은 0.5
sgd = SGD(learning_rate=0.5)

# 손실 함수(loss function)은 평균제곱오차 mse를 사용
model.compile(optimizer=sgd ,loss='mse')

# 주어진 x_data와 y데이터에 대해서 오차를 최소화하는 작업을 10번 시도
model.fit(x_data,y_data, batch_size=1, epochs=10)

# 데이터 분포와 추정된 직선 그리기
plt.plot(x_data, y_data, 'ro') # 데이터 분포
plt.plot(x_data, model.predict(x_data)) # 직선
plt.xlabel('x') # x 축 라벨
plt.ylabel('y') # y축 라벨
plt.show()
```
