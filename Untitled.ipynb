```python
from keras.datasets.mnist import load_data
# MNIST  
(trainX, trainY), (testX, testY) = load_data()
#  MNIST    
print('Train: X=%s, y=%s' % (trainX.shape, trainY.shape))
print('Test: X=%s, y=%s' % (testX.shape, testY.shape))
import matplotlib.pyplot as plt
#  25   
for i in range(25):
    # 5 x 5  
    plt.subplot(5, 5, i+1)
    #  () 
    plt.imshow(trainX[i], cmap=plt.get_cmap('gray'))
    #   
    plt.show()
```

```python
#   
#     10  
print(trainX[0][10])
# 0-255   0-1   
trainX = trainX.astype('float32') #  
testX = testX.astype('float32')
trainX = trainX / 255
testX = testX / 255
#     10  
print(trainX[0][10])
```

```python
from tensorflow.keras.utils import to_categorical
#  (one-hot encoding)
trainY = to_categorical(trainY, 10)
testY = to_categorical(testY, 10)
print(trainY)
```

```python
#     
from keras.models import Sequential
from keras.layers import Dense,Flatten
model = Sequential([ Flatten(input_shape=(28, 28)),
Dense(128, activation='relu'),
Dense(10, activation='softmax') ])

#   
model.summary()
```

```python
#   
from tensorflow.keras.optimizers import SGD
#  ,    
model.compile(optimizer='SGD', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
# 128  , 200   
model.fit(trainX, trainY, batch_size=128, epochs=200, verbose=1)
#  
test_loss, test_acc = model.evaluate(testX, testY)
print('Test accuracy: ', test_acc)
```

```python
import numpy as np
#  
image0 = trainX[0] # 0 
yhat = model.predict(np.array([image0]))
print(image0.shape, np.array([image0]).shape)
print('Predicted: class=%d' % np.argmax(yhat))
image5 = trainX[5] # 5 
yhat = model.predict(np.array([image5]))
print('Predicted: class=%d' % np.argmax(yhat))
```
